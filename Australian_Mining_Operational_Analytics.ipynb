```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from google.colab import files

# Project: WA Mining Performance 2025-2026
np.random.seed(42)
rows = 2000

data = {
    'Site_Name': np.random.choice(['Pilbara Iron', 'Kalgoorlie Gold', 'Greenbushes Lithium', 'Hunter Coal', 'Olympic Dam'], rows),
    'Date': [datetime(2025, 1, 1) + timedelta(days=np.random.randint(0, 450)) for _ in range(rows)],
    'Commodity': np.random.choice(['Iron Ore', 'Gold', 'Lithium', 'Coal', 'Copper'], rows),
    'Production_Tons': np.random.uniform(5000, 50000, rows).round(2),
    'Operational_Cost_AUD': np.random.uniform(100000, 500000, rows).round(2),
    'Safety_Incidents': np.random.choice([0, 1, 2], rows, p=[0.95, 0.04, 0.01]), # 95% days are safe
    'Equipment_Uptime_Pct': np.random.uniform(85, 99, rows).round(2)
}

df_mining = pd.DataFrame(data)
df_mining['Revenue_AUD'] = (df_mining['Production_Tons'] * np.random.uniform(80, 150)).round(2)

# Save for GitHub
df_mining.to_csv('wa_mining_data.csv', index=False)
print("✅ WA Mining Dataset Created!")

# This will trigger a download in your browser
files.download('wa_mining_data.csv')

print("🚀 Download started! Check your downloads folder.")
```
